```python
%matplotlib widget
import matplotlib.pyplot as plt
import numpy as np
```

```python
no_of_plots = 2
```

```python
cols = 3
rows = int(no_of_plots/cols) + (no_of_plots%cols!=0)
print(cols,rows)
```

```python
fig, axs = plt.subplots(rows,cols,sharex=True,sharey=True)
axs = axs.flatten()
print(axs.shape)
for i in range(no_of_plots):
    im = np.random.rand(100,100)
    axs[i].imshow(im)
for i in range(cols*rows-no_of_plots):
    fig.delaxes(axs[-(i+1)])
print(axs.shape)
```
